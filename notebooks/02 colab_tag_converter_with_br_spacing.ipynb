```python
# =====================================================
# CEL 1 — Basis: bestanden uploaden
# =====================================================

from google.colab import files

print("Upload je bronbestand (bijv. input.txt) met de originele markup.")
uploaded = files.upload()

# Neem de eerste geüploade filename als input
input_filename = list(uploaded.keys())[0]
print(f"Gebruikt invoerbestand: {input_filename}")

# Kies een naam voor de output
output_filename = "output_converted.txt"
print(f"Outputbestand wordt: {output_filename}")
```

```python
# =====================================================
# CEL 2 — Definieer de vervangingen
# =====================================================

replacements = {
    "<body>": "",
    "</body>": "",
    "<subhead_lead>": '<h5 class="Heading_heading__okScq Heading_heading--lg__v_Lob heading_lg__W3ya6" data-testid="infoblock-headline">',
    "</subhead_lead>": "</h5>",
    "<subhead>": '<h6 class="Heading_heading__okScq Heading_heading--sm__bGPWw heading_infoblockSubheading__Ecn_I heading_sm__u3F2n" data-testid="infoblock-heading">',
    "</subhead>": "</h6>",
    "<howto_facts>": '<p class="Paragraph_paragraph__exhQA Paragraph_paragraph--default-sm-default__jy0uG articleParagraph">',
    "</howto_facts>": "</p><br>",
}
```

```python
# =====================================================
# CEL 3 — Verwerken van de tekst
# =====================================================

import re

# Lees de originele inhoud
with open(input_filename, "r", encoding="utf-8") as f:
    content = f.read()

# Pas alle primaire vervangingen toe
for old, new in replacements.items():
    content = content.replace(old, new)

# Extra stap: als een paragraafblok gevolgd wordt door een h5-headline,
# voeg dan een extra <br> toe voor extra witruimte.
#
# We zoeken specifiek naar </p><br> (van howto_facts-sluiters) gevolgd door een h5-headline.
content = re.sub(
    r'(</p><br>\s*)(<h5 class="Heading_heading__okScq Heading_heading--lg__v_Lob heading_lg__W3ya6" data-testid="infoblock-headline">)',
    r'\1<br>\2',
    content
)

# Schrijf de geconverteerde inhoud naar het outputbestand
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(content)

print("Conversie voltooid.")
```

```python
# =====================================================
# CEL 4 — Resultaat bekijken en downloaden
# =====================================================

# Toon een stukje van de output in Colab
print("Voorbeeld van de geconverteerde output:\n")
print(content[:1000])  # eerste 1000 tekens tonen

# Bied het bestand aan als download
from google.colab import files
files.download(output_filename)
```
